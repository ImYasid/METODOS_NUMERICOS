```python
%load_ext autoreload
import numpy as np
import matplotlib.pyplot as plt
import math
from src import ODE_euler, ODE_euler_nth
```

### EJERCICIO UNO: METODO DE EULER VS SOLUCION REAL

```python
y_der = lambda t, y: y - t**2 + 1
y_init = 0.5

ys1a, ts1a, h = ODE_euler(a=0, b=2, f=y_der, y_t0=y_init, N=10)

print(f"El valor de h es: {h}")
print(f"Los valores de t es: {ts1a}")
print(f"Los valores de y es: {ys1a}")

def y_real(t):
    t = np.array(t)
    return (t + 1)**2 - 0.5 * np.exp(t)

ys2a = y_real(ts1a)

plt.plot(ts1a, ys1a, marker="o", linestyle="-", label="Metodo de Euler", color="GREEN")
plt.plot(ts1a, ys2a, linestyle="--", label="Solucion real", color="orange")
plt.xlabel("t")
plt.ylabel("y")
plt.title("Solución de la EDO")
plt.legend()
plt.grid(True)
plt.show()
```

### EJERCICIO TRES: Nueva funcion

```python
y_der = lambda t, y: (y/t)-(y/t)**2
y_init = 1

ys1a, ts1a, h = ODE_euler(a = 1, b = 2, f = y_der, y_t0 = y_init, N = 20)

def y_real(t):
    t = np.array(t)
    return t / (1 + np.log(t))

ys2a = y_real(ts1a)

print(f"El valor de h es: {h}")
print(f"Los valores de t es: {ts1a}")
print(f"Los valores de y es: {ys1a}")

plt.plot(ts1a, ys1a, marker="o", linestyle="-", label="Metodo de Euler", color="GREEN")
plt.plot(ts1a, ys2a, linestyle="--", label="Solucion real", color="orange")
plt.xlabel("t")
plt.ylabel("y")
plt.title("Solución de la EDO")
plt.legend()
plt.show()
```

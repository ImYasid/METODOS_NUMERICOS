```python
from scipy.optimize import newton

def f(x, points=None, show=False):
    y = x**3 - 3 * x**2 + x - 1
    if points is not None:
        points.append((x, y))
    if show:
        print(f"f({x}) = {y}")
    return y

def f_prime(x):
    return 3 * x**2 - 6 * x + 1

# M√©todo de Newton con almacenamiento de puntos intermedios
def newton_method_with_tracking(f, f_prime, x0, maxiter=50, show=False):
    points = []  # Lista para almacenar puntos intermedios

    def wrapped_f(x):
        return f(x, points=points, show=show)

    root = newton(wrapped_f, x0, fprime=f_prime, maxiter=maxiter, full_output=False, disp=False)
    return root, points

# Ejemplo de uso para el m√©todo de Newton
x0 = -1
root_newton, points_newton = newton_method_with_tracking(f, f_prime, x0, show=True)
print("\nRa√≠z encontrada (Newton):", root_newton)
print("Puntos intermedios (Newton):", points_newton)
```
